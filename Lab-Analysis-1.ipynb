```python
import pandas as pd #importing libraries
import numpy as np
```

```python
file1 = pd.read_excel('files_for_lab/excel_files/file1.xlsx') #reading files as dataframes
file2 = pd.read_excel('files_for_lab/excel_files/file2.xlsx')
file3 = pd.read_excel('files_for_lab/excel_files/file3.xlsx')
```

```python
display(file1.shape, file2.shape, file3.shape) #show shapes
```

```python
display(file1.columns, file2.columns, file3.columns) #check columns
```

```python
#lowering cases
cols = [] 
for i in file1.columns: cols.append(i.lower())
file1.columns = cols   

cols = []
for i in file2.columns: cols.append(i.lower())
file2.columns = cols

cols = []
for i in file3.columns: cols.append(i.lower())
file3.columns = cols  

#replacing names
file1 = file1.rename(columns={'st':'state','customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto','number of open complaints':'number_of_open_complaints','vehicle class':'vehicle_class', 'total claim amount':'total_claim_amount'}) 
file2 = file2.rename(columns={'st':'state','customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto','number of open complaints':'number_of_open_complaints','vehicle class':'vehicle_class', 'total claim amount':'total_claim_amount'})
file3 = file3.rename(columns={'st':'state','customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto','number of open complaints':'number_of_open_complaints','vehicle class':'vehicle_class', 'total claim amount':'total_claim_amount'})

#changing order
file1 = file1[['customer', 'state', 'gender', 'education', 'income', 'customer_lifetime_value', 'monthly_premium_auto', 'policy type', 'vehicle_class', 'number_of_open_complaints', 'total_claim_amount']]
file2 = file2[['customer', 'state', 'gender', 'education', 'income', 'customer_lifetime_value', 'monthly_premium_auto', 'policy type', 'vehicle_class', 'number_of_open_complaints', 'total_claim_amount']]
file3 = file3[['customer', 'state', 'gender', 'education', 'income', 'customer_lifetime_value', 'monthly_premium_auto', 'policy type', 'vehicle_class', 'number_of_open_complaints', 'total_claim_amount']]
```

```python
(file1.columns == file2.columns) & (file3.columns == file2.columns) #check if columns are equal in all dataframes
```

```python
data = pd.concat([file1, file2, file3], axis=0) #concatenate data
data.shape
```

```python
data.info()
```

```python
data._get_numeric_data() #numerical columns
```

```python
data.select_dtypes(include=['object']) #categorical columns
```

```python
data.info()
```

```python
data_clean = data.drop(['education', 'number_of_open_complaints'], axis=1) #remove education and complaints columns
data_clean.head()
```

```python
data_clean['customer_lifetime_value'] = pd.to_numeric(data_clean['customer_lifetime_value'],errors = 'coerce') #conver to numbers
data_clean['customer_lifetime_value'] = data_clean['customer_lifetime_value']*100 #multiply by 100
data_clean
```

```python
data_clean[data_clean.duplicated() == True] #check for duplicates
```

```python
data_clean = data_clean.drop_duplicates(subset=None,keep='first') #drop duplicates
```
